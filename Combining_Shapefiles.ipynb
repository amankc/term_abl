# This section merges two shapefiles

```python
import geopandas as gpd
import pandas as pd
import osgeo as osg
from osgeo import ogr
import os
```

```python
glacier_name = 'Sermeq_Kujalleq'
glacier_id = 'SEK'
root_path = '/Users/amankc/Terminus_Ablation/'
shp_path = root_path + glacier_name + '/Terminus_Positions/' + glacier_name + '.shp'
folder_name = '/Users/amankc/Library/CloudStorage/GoogleDrive-amankc929@gmail.com/My Drive/Terminus'
add_shp_path = folder_name + '/Terminus_'+ glacier_id +'.shp'
# add_shp_path = folder_name + '/Terminus_'+ glacier_id + ' (2)'+'.shp'
# folder_name = '/Users/amankc/Library/CloudStorage/GoogleDrive-amankc929@gmail.com/My Drive/Terminus/GEEDit-delineations/'
# add_shp_path = folder_name + glacier_id+ '/Terminus_'+ glacier_id +'.shp'
# add_shp_path = folder_name +glacier_id+'/Terminus_'+ glacier_id + ' (2)'+'.shp'
shp_1 = gpd.read_file(shp_path)
shp_2 = gpd.read_file(add_shp_path)
# Bring the shapefiles into common cordinate system
shp_1 = shp_1.to_crs('EPSG:3413')
shp_2 = shp_2.to_crs('EPSG:3413')
# # # Merge/Combine multiple shapefiles into one
combined_shp = gpd.pd.concat([shp_1, shp_2])
 
# #Export merged geodataframe into shapefile
combined_shp.to_file(root_path + glacier_name + '/Terminus_Positions/'+'Combined_' + glacier_name + '.shp')
```
